Verificamos que o t test ficou semelhante ao z test

Mas lembre-se:


Use o t-test quando:

O tamanho da amostra é pequeno (n < 30).
O desvio padrão populacional é desconhecido.
A distribuição da população é aproximadamente normal.
Use o z-test quando:

O tamanho da amostra é grande (n ≥ 30).
O desvio padrão populacional é conhecido.
A população é normalmente distribuída ou a amostra é suficientemente grande para aplicar o teorema do limite central.



**Teste z (z-test)**

Tamanho da amostra grande: O z-test é **usado quando o tamanho da amostra é maior que 30.**

**Desvio padrão populacional conhecido**: O desvio padrão da população é conhecido ou a amostra é suficientemente grande para que a distribuição da amostra seja aproximadamente normal (teorema do limite central).

**Distribuição normal:** O z-test assume que a população segue uma distribuição normal.

Existem dois tipos principais de z-tests:

**z-test de uma amostra:** Compara a média de uma amostra com um valor conhecido.

**z-test de duas amostras:** Compara as médias de duas amostras independentes.

**Teste t (t-test)**

Tamanho da amostra pequeno: Geralmente, o t-test **é usado quando o tamanho da amostra é menor que 30**.

**Desvio padrão populacional desconhecido**: Quando o desvio padrão da população não é conhecido e, portanto, deve ser estimado a partir da amostra.

**Distribuição aproximadamente normal**: A amostra deve vir de uma população que se aproxima de uma distribuição normal.

Existem três tipos principais de t-tests:

**t-test de uma amostra**: Compara a média de uma amostra com um valor conhecido.

**t-test de duas amostras independentes:** Compara as médias de duas amostras independentes.

**t-test pareado:** Compara médias de duas amostras dependentes (pares correspondentes de observações).

Agora vamos fazer um teste para verificar se a média de notas do filme Toy Story é 3.42205.

```python
p_valor > 0.05
```

Vamos plotar diversas notas do filme Toy Story

Queremos ver visualmente como a quantidade de amostras afeta a média e o intervalo de confiança. Ou seja, queremos comprovar o por que devemos utilizar Z test para amostras maiores que 30 e T test para quantidade menores.

O código está realizando uma análise para verificar como as médias de rating e os valores p do teste z mudam à medida que o número de observações aumenta. A linha vermelha em y = 0.05 é um limiar típico para significância estatística, onde valores p abaixo de 0.05 indicam que a média dos ratings é significativamente diferente de 3.42205.

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating


## A função para calcular  a média e o teste z para os elementos de temp
def calcular_teste(i):
  media = temp[0:i].mean()
  z, p_valor = ztest(temp[0:i], value = 3.42205)
  return (i, media, p_valor)


## Cria uma lista de tuplas contendo o índice, a média e o valor p para cada subconjunto de temp desde o índice 2 até o tamanho total de temp. Esta lista é convertida para um array NumPy.
medias = np.array([calcular_teste(i) for i in range(2, len(temp))])

plt.plot(medias[:,0],medias[:,1])
plt.plot(medias[:,0],medias[:,2])
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors = 'red')
plt.show()
```

Vamos utilizar o Ranksums para então realizar uma análise estatística não paramétrica.

O Ranksums é baseado no Wilconxon, de forma rankeada

```python
from scipy.stats import ranksums
```

```python
ranksums(notas1.rating, notas593.rating)
```

```python
stat, p_valor = ranksums(notas1.rating, notas593.rating)
```

```python
print('stat =', stat)
print('p_valor =', p_valor)
```

hipótese nula é de que as duas amostras vieram da mesma distribuição. Já a hipótese alternativa, que admitiremos nesse caso (já que o p value foi bem pequeno), é a de que os valores de uma amostra têm uma tendência a serem maiores que os da outra amostra.

Repare que, quando trabalhamos com Teste T e Teste Z estávamos nos referindo às médias - o que fazia sentido em uma distribuição normal. Porém, em uma distribuição não paramétrica, não usamos as médias, mas sim amostras aleatórias de cada conjunto testado.

Ou seja, se pegarmos notas aleatórias para Toy Story e Silence of the Lambs, as notas deste último, em geral, são maiores. Análises como essa, junto com as médias, os boxplots, os histogramas e outras informações que aprendemos a testar anteriormente, nos permitem concluir quando um filme é melhor que outro.

**Double dipping**: quando você mergulha sua mão duas vezes no mesmo dado, uma para levantar uma hipótese e outra para validá-la. Isso é perigoso, pois muitas vezes a hipótese vai se validar, uma vez que você usou os próprios dados para levantá-la.

Como evitar essa situação? Basta validarmos a hipótese num conjunto diferente de dados de onde levantamos ela. Isto é, temos que, de antemão, separar nossos dados em mais de uma parte, levantar hipóteses e modelos em uma parte e validá-los em outra.
